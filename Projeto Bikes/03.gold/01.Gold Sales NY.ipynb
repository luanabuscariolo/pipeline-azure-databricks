```sql
%python
bronze_path   = 'abfss://data@datalakedatabrickstest.dfs.core.windows.net/bikestore/bronze/'
silver_path   = 'abfss://data@datalakedatabrickstest.dfs.core.windows.net/bikestore/silver/'
gold_path     = 'abfss://data@datalakedatabrickstest.dfs.core.windows.net/bikestore/gold/'
resource_path = 'abfss://data@datalakedatabrickstest.dfs.core.windows.net/bikestore/resource/origem/'
resource_path_volume = '/Volumes/bikestore/logistics/bikestore_resource/origem/'
```

```sql
%python
silver_map = {
    "tmp_silver_customer":      f"{silver_path}/customer/",
    "tmp_silver_orders":        f"{silver_path}/orders/",
    "tmp_silver_product":       f"{silver_path}/product/",

}
for view_name, path in silver_map.items():
    (spark.read.format('delta')
        .load(path)
        .createOrReplaceTempView(view_name))
 
```

```sql
select * from tmp_silver_customer limit 4
```

```sql
select * from tmp_silver_product limit 4
```

```sql
select 
  shipped_date
  ,round(sum(total_sale),2) as total_sale
  --,state,status
 from tmp_silver_orders 
 where state = 'NY'
 and status = 'Delivered'
 and shipped_date is not null
 group by shipped_date
 --,state  ,status
```

```sql
%python
df_sales_ny_gold = spark.sql("""
                               
select 
  shipped_date
  ,round(sum(total_sale),2) as total_sale
  --,state,status
 from tmp_silver_orders 
 where state = 'NY'
 and status = 'Delivered'
 and shipped_date is not null
 group by shipped_date
 --,state  ,status
         
                              
                              """)

# salvar em Delta na gold
df_sales_ny_gold.write\
    .mode('overwrite')\
    .format('delta')\
    .option('mergeSchema','true')\
    .save(f'{gold_path}/sales_ny')
    
```
